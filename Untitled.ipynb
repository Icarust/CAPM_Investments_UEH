```python
import numpy as np
import pandas as pd
import Input_Inv as Input

%load_ext autoreload
%autoreload 2
```

```python
tickers = ['CTG.csv','MBB.csv','TCB.csv','TPB.csv','VCB.csv','VNIndex.csv','MWG.csv','FRT.csv','SAS.csv','TLP.csv','VGC.csv']
port = pd.DataFrame()
for t in tickers:
    port[t] = Input.read_data(t)['Close']
```

```python
port
```

```python
port.columns = ['CTG','MBB','TCB','TPB','VCB','VNIndex','MWG','FRT','SAS','TLP','VGC']
```

```python
port
```

```python
port_return = np.log(port/port.shift(1))
```

```python
port_return
```

```python
# Beta =  Covariance (CTG/VNIndex)/Var(VNIndex)
```

```python
port_Cov = port_return.cov()*252
```

```python
port_Cov
```

```python
port_Corr = port_return.corr()
port_Corr
```

```python
ones_Corr = np.ones_like(port_Corr,dtype=bool)
ones_Corr
```

```python
mask = np.triu(ones_Corr)
mask
```

```python
import seaborn as sns
```

```python
adjust_mask = mask[1:, :-1]
adjust_mask
```

```python
adjust_port_Corr = port_Corr.iloc[1:, :-1]
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(10,8))

cmap = sns.diverging_palette(0, 230, 90, 60, as_cmap=True)

sns.heatmap(data=adjust_port_Corr, mask=adjust_mask, annot=True, cmap=cmap,linecolor='White',linewidths=0.1)

title = 'Tương quan giữa từng loại tài sản trong danh mục\n'
ax.set_title(title, loc='center',fontsize=20)
```
